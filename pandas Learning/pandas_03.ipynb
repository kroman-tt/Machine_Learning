### Getting some basic info about data

### Distribution plot

```python
# Histrogram
sns.histplot(data =  iris, x = 'petal_width', bins = 20)
plt.title("Distrubution of Petal Width")
plt.show()
```

```python
# Histrogram
sns.histplot(data =  iris, x = 'petal_width', hue = 'species', bins = 20)
plt.title("Distrubution of Petal Width")
plt.show()
```

```python
# Barplot

sns.barplot(data = iris, x = 'species', y = 'sepal_length', palette = 'Set2')
plt.title("BAr graph distribution of sepal_length based on species")
```
