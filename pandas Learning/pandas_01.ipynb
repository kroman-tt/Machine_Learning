```python
student_dict = {
    "student": ["student1","student2","student1","samir","student5","student6","student7","student8"],
    "Roll No": [1,np.nan, 1,3,4,5,6,7],
    "Gender": ["Male","Female","Male",np.nan,"Male","Female","Female","Male"],
    "Marks_obtained":[56,76,56,43,67, np.nan ,73,75]
}
```

```python
student_df = pd.DataFrame(student_dict)
student_df
```

### To check the duplicate value

```python
student_df.duplicated()
```

```python
# (.sum()) gives the value of duplication
student_df.duplicated().sum()  
```

```python
# Now deleting the duplicate value 
student_df = student_df.drop_duplicates()
student_df
```

```python
#deleting the null value records by using (.dropna)
student_df2 = student_df.dropna()
student_df2
```

### Tmputation method:(mean,median and mode)

```python
gender_mode = student_df["Gender"].mode()
gender_mode
```

```python
#[0] give the 0 index value from above 
gender_mode = student_df["Gender"].mode()[0]
gender_mode
```

```python
student_df["Gender"] = student_df["Gender"].fillna(gender_mode)
student_df
```

```python
marks_mean = student_df["Marks_obtained"].mean()
```

```python
marks_mean = np.round(marks_mean, 2)
marks_mean
```

```python
student_df["Marks_obtained"] = student_df["Marks_obtained"].fillna(marks_mean)
student_df
```

```python
student_df = student_df.dropna()
student_df
```
