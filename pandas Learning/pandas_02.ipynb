### Some feature engineering


```python
df["Familysize"] = df["SibSp"] + df["Parch"] + 1
```

```python
df.head()
```

```python
df = df.drop(columns = [ "SibSp" , "Parch"])
```

```python
df.head()
```

### Handaling the missing value

```python
## For age 
mean_age = df["Age"].mean()
mean_age = np.round(mean_age, 2)

# mean_age

# for embarked
mode_embark = df["Embarked"].mode()[0]
mode_embark
```

```python
df["Embarked"].value_counts
```

```python
mode = df["Embarked"].mode() [0]
mode
```

```python
df['Age'] = df["Age"].fillna(mean_age)
```

```python
df["Embarked"] = df["Embarked"].fillna(mode)
```

```python
df.isnull().sum()
```

```python
df["HasCabin"] = np.where(df["Cabin"].notna(), "Yes" , "No")
```

```python
df.head()
```

```python
df = df.drop(columns= ["Cabin"])
```

```python
df.head()
```

### Handaling categorical features (columns)

```python
df.head()
```

```python
df["Sex"] = df["Sex"].map({"male": 1 , "female":0})
```

```python
df.head()
```

```python
df["HasCabin"] = df["HasCabin"].map({"Yes":1 , "No":0 })
```

```python
df.head()
```

```python
df["Embarked"] = df["Embarked"].map({"S":0 , "C":1 , "Q":2 })
```

```python
df.head()
```
