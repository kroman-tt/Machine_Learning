```python
# !pip install seaborn

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
iris = pd.read_csv(r"C:\Users\LENOVO\Desktop\Machine learning\IRIS (2).csv")
```

```python
iris.head()
```

### Getting the  basic info about the data

```python
iris.info()
```

### Distribution plot

```python
# Histogram
sns.histplot(data = iris, x = 'petal_width', hue = 'species', bins = 20)
plt.title("Distribution of Petal width")
plt.show()
```

```python
sns.histplot(data = iris, x = 'sepal_length', bins = 20, hue = 'species')
plt.title("Distribution of sepal length")
plt.show()
```

```python
## Barplot

sns.barplot(data = iris, x = 'species', y = 'sepal_length', palette = 'Set2')
plt.title("Bar graph distribution of sepal_length based on species")
plt.show()
```

```python
sns.barplot(data = iris, x = 'species', y = 'sepal_width', palette = 'Set2')
plt.title("Bar graph distribution of sepal_width based on species")
plt.show()
```

```python
iris.head(2)
```

```python
# Scatter plot

sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python
sns.scatterplot(data = iris, x = 'sepal_width', y = 'petal_length', hue = 'species')
plt.show()
```

```python
## Boxplot
sns.boxplot(data = iris, x = 'species', y = 'sepal_length')
plt.title("Boxplot showing the distribution of sepal length")
```
