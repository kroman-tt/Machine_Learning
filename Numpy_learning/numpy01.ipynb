```python
import numpy as np
```

### Learning array in numpy

```python
### 1d array(one dimensoin array)

array_1d = np.array([1,2,3,4,5,6])
array_1d
```

### Two dimensional array

```python
array_2d = np.array([[2,3,4],[1,2,3],[5,6,7]])
print(array_2d)
```

### Three dimentional array
-->collection of 2d array

```python
array_3d =   np.array([[[1,2,3],
                        [2,4,5],
                        [4,5,6]],
                        [[2,3,4],
                        [5,6,7],
                        [8,9,7]]])
print(array_3d)
```

### To check the dimension of array

```python
array_3d.ndim
```

```python
array_2d.ndim
```

```python
array_1d.ndim
```

### One dimension array

```python
array_1d = np.arange(1,101)
print(array_1d)
```

```python
array_1d = np.arange(1,101,2)
print(array_1d)
```

### Linespace method

```python
array_1d = np.linspace(1,10,10)
print(array_1d)
```

```python
array_1d = np.linspace(1,5,10)
print(array_1d)
```
