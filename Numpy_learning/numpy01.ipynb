### concatination in numpy array


```python
array1 = np.array([1,2,3])
array2 = np.array([1,2,4])
array3 = np.concatenate((array1,array2))
print(array3)
```

### For two dimentional array

```python
arr2d_1 = np.array([[1,2,3,4],
                    [5,6,7,8],
                    [6,9,7,3]])
arr2d_2 = np.array([[1,5,3,4],
                    [5,6,9,8],
                    [6,1,7,3]])
array3 = np.concatenate((arr2d_1,arr2d_2), axis=0)
print(array3)
```

```python
arr2d_1 = np.array([[1,2,3,4],
                    [5,6,7,8],
                    [6,9,7,3]])
arr2d_2 = np.array([[1,5,3,4],
                    [5,6,9,8],
                    [6,1,7,3]])
array3 = np.concatenate((arr2d_1,arr2d_2), axis=1)
print(array3)
```

### hstack (h= horizontal) and vstack (v= vertical) concatenation

```python
arr1 = np.array([[1,2,3,1],[4,5,8,9]])
arr2 = np.array([[2,3,4,5]])
arr3 = np.vstack((arr1,arr2))
print(arr3)
arr4 = np.array([[4],
                [3]])
```

```python
print(np.hstack((arr1,arr4)))
```

### mean


```python
array1 = np.array([12,14,16,13,17])
mean = np.mean(array1)
print(mean)
```

```python
median = np.median(array1)
print(median)
```

```python
std = np.std(array1)
print(std)
```

```python
array_2 = np.random.randint(15,25, size=(6,4))
print(array_2)
```

```python
array_2_mean = np.mean(array_2, axis=0)
print(array_2_mean)
```

```python
array_2_mean = np.mean(array_2, axis=1)
print(array_2_mean)
```
